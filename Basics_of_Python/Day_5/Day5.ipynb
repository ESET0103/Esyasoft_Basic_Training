```python
x = [0, 2, 4, 6, 8]
y = [0, 4, 16, 36, 64]
```

```python
ax = plt.subplot()
ax.plot(x, y, marker='o', label='Data Points')
ax.set_title('Basic Components of Matplotlib Figure')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
plt.show()
```

### Combined Ecample

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = np.array([10,20,30])
```

```python
df = pd.DataFrame({"Numbers":arr})
```

```python
plt.bar(df.index, df["Numbers"])
plt.show()
```

```python
dict = {'A': 10, 'B': 20, 'C': 30}
plt.bar(dict.keys(), dict.values())
plt.title('Bar Chart from Dictionary')
plt.xlabel('Keys')
plt.ylabel('Values')
colors = ['red', 'blue', 'green']
plt.bar(dict.keys(), dict.values(), color=colors)
plt.show()
```

### Quantile ranges Q1, Q2, Q3, Q4
### Q1 = df.quantile(0.25)
### Q3 = df.quantile(0.75)
### IQR = Q3 - Q1 

```python
dff = df.select_dtypes(include=['number'])
Q1 = dff.quantile(0.25)
Q3 = dff.quantile(0.75)
IQR = Q3 - Q1 
print(IQR)
```
